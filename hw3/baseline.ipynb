```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('train.csv')
train_end_idx = len(df) 
df_test = pd.read_csv('test.csv')
df_test['RainToday'] = np.zeros((len(df_test),))

# 為了處理方便，把 'train.csv' 和 'test.csv' 合併起來，'test.csv'的 RainToday 欄位用 0 補起來。
# 以 train_end_idx 作為 'train.csv' 和 'test.csv' 分界列，
df = pd.concat([df, df_test], sort=False)
```

```python
df['Date'] = pd.to_datetime(df['Date'], 
 format = '%Y/%m/%d', 
 errors = 'coerce')

df.info()
```

```python
# Standardize

from sklearn.preprocessing import StandardScaler
df_new = df
scaler = StandardScaler().fit(df_new)

new = scaler.transform(df_new)
df = pd.DataFrame(new, columns = df.columns)
df
```

```python
# 將非數值欄位拿掉
#df = df.drop(columns = [col for col in df.columns if df[col].dtype == np.object])
#Label encoding (Date)
labelencoder = LabelEncoder()
df['Date'] = labelencoder.fit_transform(df['Date'])
#One-hot encoding (other non-value object)
df = pd.get_dummies(df)
df

#df.fillna(0)
#for column in list(df.columns[df.isnull().sum() > 0]):
#  mean_val = df[column].mean()
#  df[column].fillna(mean_val, inplace=True)
```
